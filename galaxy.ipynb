````python
from random import randint
from statistics import mean, median
````

````python
from matplotlib import pyplot, rcParams
````

````python
v = []
means = []
medians = []
y = []
for i in range(10000):
    v.append([])
    for j in range(randint(3, 1000)):
        v[i].append(randint(-100, 100))
    
    means.append(mean(v[i]))
    medians.append(median(v[i]))
    y.append((max(v[i]) + min(v[i])) / 2)
````

````python
rcParams["figure.figsize"] = 15, 15
pyplot.scatter(means, y, marker="x", color="xkcd:baby blue")
pyplot.xlabel("mean")
pyplot.ylabel("idk")
#pyplot.xlim([-4, 7])
#pyplot.ylim([-4, 7])
pyplot.gca().set_facecolor("black")
pyplot.show()

pyplot.scatter(medians, y, marker="x", color="xkcd:baby blue")
pyplot.xlabel("median")
pyplot.ylabel("idk")
pyplot.gca().set_facecolor("black")
#pyplot.xlim([-4, 7])
#pyplot.ylim([-4, 7])
pyplot.show()
````
