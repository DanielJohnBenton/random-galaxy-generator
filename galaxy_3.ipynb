```python
from random import randint, random
from statistics import mean, median
```

```python
from matplotlib import pyplot, rcParams
```

```python
def vectors(minAmount, maxAmount):
    v = []
    means = []
    medians = []
    y = []
    yDiv = randint(-5, 5) if randint(1, 2) == 1 else random()
    if yDiv == 0:
        yDiv = randint(1, 5) if randint(1, 2) == 1 else randint(-5, -1) if randint(1, 2) == 1 else random()

    yDivChanges = randint(1, 2) == 1

    for i in range(randint(minAmount, maxAmount)):
        v.append([])
        for j in range(randint(3, 1000)):
            v[i].append(randint(-100, 100))

        means.append(mean(v[i]))
        medians.append(median(v[i]))

        if yDivChanges:
            yDiv += randint(-5, 5) if randint(1, 2) == 1 else random() if randint(1, 2) == 1 else random() * -1
            if yDiv == 0:
                yDiv = random()

        y.append((max(v[i]) + min(v[i])) / yDiv)
    
    return means, medians, y

def choose_marker():
    if randint(1, 100) <= 10:
        return "x"
    if randint(1, 100) <= 30:
        return "o"
    return "."

def generate_galaxy():
    markers = ["x", ".", "o"]
    colours = ["xkcd:baby blue", "xkcd:sky blue", "xkcd:sunshine yellow", "xkcd:light yellow", "xkcd:cyan", "xkcd:white", "xkcd:sea green"]
    
    means, medians, y = vectors(5000, 10000)
    
    meansOrMedians = "means" if randint(1, 100) <= 30 else "medians"
    x = means if meansOrMedians == "means" else medians
    
    xy = randint(1, 100)

    rcParams["figure.figsize"] = 10, 10
    pyplot.scatter(x if xy <= 85 else y, y if xy <= 85 else x, marker=choose_marker(), color=colours[randint(0, len(colours) - 1)])
    
    for i in range(1, 3):
        if randint(1, 100) <= (85 if i == 1 else 60):
            means, medians, y = vectors(3, 3333)
            xy = randint(1, 100)
            if meansOrMedians == "means":
                x = means if randint(1, 100) <= 70 else medians
            else:
                x = medians if randint(1, 100) <= 70 else means
            pyplot.scatter(x if xy <= 75 else y, y if xy <= 75 else x, marker=choose_marker(), color=colours[randint(0, len(colours) - 1)])
            
    
    pyplot.xticks([], [])
    pyplot.yticks([], [])
    pyplot.gca().set_facecolor("black")
    pyplot.show()
```

```python
for i in range(150):
    generate_galaxy()
```
